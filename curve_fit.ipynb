```python
%matplotlib inline
import matplotlib.pyplot as plt #module dari matplotlib yang digunakan untuk menggambar grafik
import numpy as np
from sklearn.pipeline import make_pipeline #memanggil library untuk membuat pipeline
from sklearn.preprocessing import PolynomialFeatures #kelas di sciki-learn untuk mengubah fitur data menjadi polinomial
from sklearn.linear_model import LinearRegression #kelas di sciki-learn untuk membuat model regresi linear
```

```python
rng = np.random.RandomState(1) #memnuat randomstate untuk menghasilkan angka acak
x1 = 10 * rng.rand(50) #menghasilkan 50 angka acak dengan rentang 1-10
y1 = 2 * x1 - 5 + rng.randn(50) #menghasilkan sedikit gangguan/noise
plt.scatter(x1, y1); #memvisualisasikannya
```

```python
model1 = LinearRegression()
model1.fit(x1[:, np.newaxis], y1) #fungsi fit untuk melatih model data yg ada

xfit1 = np.linspace(0, 10, 1000) #array yg berisi 1000 nilai
yfit1 = model1.predict(xfit1[:, np.newaxis]) #fungsi untuk menghitung nilai y yang diprediksi berdasarkan nilai x yang baru

plt.scatter(x1, y1) #membuat scater plot dengan data asli
plt.plot(xfit1, yfit1); #membuat garis yang menggambarkan hasil prediksi regresi linier berdasarkan model yang sudah dilatih
```

```python
print("Model slope:    ", model1.coef_[0]) #mencetak nilai seberapa besar perubahan pada y keika x berubah
print("Model intercept:", model1.intercept_) #mencetak nilai intercept dari model
```

```python
#menghasilkandata secara acak kemudian memvisualisasikannya dengan scatter plot seperti kode diatas
rng = np.random.RandomState(1)
x2 = 10 * rng.rand(50)
y2 = np.sin(x2) + 0.1 * rng.randn(50)

plt.scatter(x2, y2)
```

```python
#melatih model dengan drajat 6 untuk data x2 dan y2
model2 = make_pipeline(PolynomialFeatures(6), LinearRegression()) #mengubah data menjadi fitur polinomial lalu regresi liner akan digunakan untuk melatih model
model2.fit(x2[:, np.newaxis], y2)
xfit2 = np.linspace(0, 10, 1000)
yfit2 = model2.predict(xfit2[:, np.newaxis])

#menggambahkan garis kurva hasil prediksi regresi di atas scatter polot
plt.scatter(x2, y2)
plt.plot(xfit2, yfit2);
```

```python
#mengakses model LinearRegression yang ada dalam pipeline dan mencetak slope dan intercept
liniear_model = model2.steps[1][1]
print("Model slope:    ", liniear_model.coef_[0])
print("Model intercept:", liniear_model.intercept_)
```
